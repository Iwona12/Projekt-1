```python
"""
   Pobieranie danych ze strony www.otodom.pl. Strona otodom czesto ulega zmianie, co wymaga zmiany kodu. 
   Dane zostaly pobrane 2 maja 2019r.
"""

import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy.conf import settings
from bs4 import BeautifulSoup
import requests


class OtodomBot(scrapy.Spider):
    name = "otodom_scraper"
    def start_requests(self):
        base = "https://www.otodom.pl/sprzedaz/mieszkanie/katowice/?search%5Bdist%5D=0&search%5Bsubregion_id%5D=306&search%5Bcity_id%5D=172&page="
        for page_number in range(1, pages_all+1):
            url = base+str(page_number)
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
        urls = response.css("div.offer-item-details > header > h3 > a::attr(href)").extract()
        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse_one_page)

    def parse_one_page(self, response):
        price = response.css("#root > div > article > header > div.css-xiali8-AdHeader-className > div.css-19hg3qf-AdHeader-className > div.css-c0ipkw-AdHeader::text").extract_first()
            
        details = response.css("#root > div > article > div.css-k68zts-AdPageTemplate-className > div.css-b8lhu7-AdPageTemplate-className > section.section-overview > div.css-1v9yl6n-AdOverview-className > ul > li").extract()
        
        if price is not None:
            record = {"price": price,
                      "details": details,
                      "url": response.request.url}
            real_estates.append(record)
        

real_estates = []

#otodom_katowice = "https://www.otodom.pl/sprzedaz/mieszkanie/katowice/?search%5Bdist%5D=0&search%5Bsubregion_id%5D=306&search%5Bcity_id%5D=172&page=1"
#soup = BeautifulSoup(requests.get(otodom_katowice).text, 'html.parser')
#pages_all = int(soup.select("#pagerForm > ul > li.pager-counter > strong")[0].text)
pages_all = 2  # tylko dla celow testowych
process = CrawlerProcess()
process.crawl(OtodomBot)
process.start()

```

```python
"""
Tworzenie listy z "Price" (Ceny) i czyszczenia ze spacji, zl i kropek.
"""
Ceny = []
for element in real_estates:
    cena = element['price'].replace(" ","").replace("zł","").replace(",", ".")
    Ceny.append(cena)
Ceny   
```

```python
"""Usuwanie rekordu jezeli pozycji w Cena nie da sie przeksztalcic na float. """
for i in range(0, len(Ceny)):
    
    m = 0
    try:
        m = float(Ceny[i])
        
    except ValueError:
        Ceny[i] = None
        real_estates[i] = None
        for i in range(0, len(Ceny)):
            Ceny.remove(None)
            real_estates.remove(None)
```

```python
"""
Tworzenie listy z "details" (Dane) i czyszczenie danych.
"""

Dane = []
szczegol = []
for i in range(0, len(real_estates)):
    szczegol = []
    for j in range (0, len(real_estates[i]['details'])):
        szczegol.append(real_estates[i]['details'][j].replace("<li>","").replace(" zł", "").replace("<strong>","").replace("</strong></li>","").replace(" m²", "").replace(",", "."))
    Dane.append(szczegol)
Dane
```

```python
"""
Tworzenie slownika z danych "details" dla jednego przykladu.
"""
B = []
s = {}
for i in range(0, len(Dane[0])):
    s[Dane[0][i].split(': ')[0]] = Dane[0][i].split(': ')[1]
    
B.append(s)
B
```

```python
"""
Tworzenie slownika z danych "details".
"""

Baza_slownik = []

for i in range(0, len(Dane)):
    slownik = {}
    for j in range(0, len(Dane[i])):
        slownik[Dane[i][j].split(': ')[0]] = Dane[i][j].split(': ')[1]
    Baza_slownik.append(slownik)
Baza_slownik
```

```python
"""
Tworzenie listy z kluczy slownika z danych "details".
"""

Keys = []

for i in range(0, len(Dane)):
    for key in Baza_slownik[i].keys():
        if key not in Keys:
            Keys.append(key)

Keys
```

```python
"""
Tworzenie listy kluczy.
"""


Tabela = []
for i in range(0, len(Baza_slownik)):
    wiersz = []
    for key in Keys:
        if key not in Baza_slownik[i].keys():
            Baza_slownik[i][key] = ''
        wiersz.append(Baza_slownik[i][key])
    Tabela.append(wiersz)
    
Tabela
```

```python
"""
Czyszenie danych. Przypisanie wartosci liczbowych dla parteru, suterenu, poddasza.
"""
for i in range(0, len(Tabela)):
    Tabela[i][0] = Tabela[i][0].replace(" ", "")
    Tabela[i][4] = Tabela[i][4].replace("parter", "0").replace("suterena", "-1").replace("poddasze", "100")
    #Tabela[i][4] = Tabela[i][4].replace("parter", "0").replace("&gt; 10", "10").replace("suterena", "-1").replace("poddasze", "100")
```

```python
"""
Laczenie danych z "Tabela" i "Ceny".
"""
for i in range(0, len(Tabela)):
    Tabela[i].append(Ceny[i])
Keys.append('Cena')
Tabela
```

```python
import csv
import pandas as pd
```

```python
"""
Tworzenie DataFrame.
"""
values = Tabela
columns = Keys

Mieszkania = pd.DataFrame(values, columns = columns)
Mieszkania
```

```python
"""
Zmiana danych w "Cena" na float.

"""
Mieszkania['Cena'] = Mieszkania['Cena'].astype('float64')
```
